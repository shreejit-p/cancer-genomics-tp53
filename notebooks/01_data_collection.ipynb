```python
# Import required BioPython modules
from Bio import Entrez
from Bio import SeqIO
```

```python
# Set the Entrez email
Entrez.email = "shreejit.data@gmail.com"
```

```python
#Search for TP53 genes in humans
search_handle = Entrez.esearch(
    db = "nucleotide",
    term = "TP53[Gene] AND Homo sapiens[Organism]  AND mRNA[Filter] AND RefSeq[Filter]",
    retmax = 10
)
search_results = Entrez.read(search_handle)
search_handle.close()
# Extract list of IDs
id_list = search_results["IdList"]
print("Retrieved IDs:")
print(id_list)
```

```python
# Fetch sequences in FASTA format using retrieved IDs
fetch_handle = Entrez.efetch(db = "nucleotide",
                             id = id_list,
                             rettype = "fasta",
                             retmode = 'text')
fasta_data = fetch_handle.read()
fetch_handle.close()

# Preview first 500 characters
print(fasta_data[:500])
```

```python
# Save raw sequences to data folder
output_path = "../data/raw_sequences.fasta"
with open(output_path,"w") as file:
    file.write(fasta_data)
print("Data saved to :", output_path)
```

```python
# Parse FASTA file using SeqIO
sequences = list(SeqIO.parse("/home/dragonfly/cancer-genomics-tp53/data/raw_sequences.fasta", "fasta"))
print("Total Sequences:", len(sequences))

# Inspect sequences
for seq in sequences:
    print("ID:", seq.id)
    print("Length:", len(seq))
    print()
```
